W5A18 VIết hàm kiểm tra một số có phải số đối xứng không, xử lý trên dữ liệu kiểu số nguyên

```python
#W5A18
def check(n: int):
    so = len(str(n))
    for i in range (0,so-1):
        if str(n)[i] != str(n)[so-i-1]:
            return False
    else:
        return True
assert check(3)==True
assert check(0)==True
assert check(12)==False
assert check(12344321)==True
assert check(1991)==True
print("all test accepted")
```

```python
#W5A20
def  detect_number_type(s):
    if type(s)==int:
        return "int"
    elif type(s)==float:
        return "float"
    else:
        return "neither"
assert detect_number_type(2)=="int"
assert detect_number_type(2.0)=="float"
assert detect_number_type("a")=="neither"
assert detect_number_type(0)=="int"
assert detect_number_type("!")=="neither"
print("all test accepted")
```

```python
#W5A21
def  is_one_product_of_others(a,b,c):
    so = 0

    while (a*b)!=c:
        a,b,c=b,c,a
        so +=1
        if so == 3:
            break
    else:
        return True
    if so == 3:
        return False
assert is_one_product_of_others(1,1,1)== True
assert is_one_product_of_others(1,1,2)== False
assert is_one_product_of_others(1,2,3)== False
assert is_one_product_of_others(0,1,2)== False
assert is_one_product_of_others(1,2,1)== False
print("all test accepted")

    
```

W5A30 Viết hàm tính hệ số nhị thức C(n, k)

```python
#W5A30
def gt(n):
    gt = 1
    for i in range(1,n+1):
        gt *=i
    return gt
def C(n,k):
    return gt(n)/(gt(k)*gt(n-k))
assert C(12,6)==924
assert C(20,18)==190
assert C(50,1)==50
assert C(20,20)==1
print ("all test accepted")
```
