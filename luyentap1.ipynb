```python
x,y,z= list(map(int,input().split()))
#true/false
print("{0}<{1} la {2}".format(x,y,x<y))
print("{0}!={1} la {2}".format(x,y,x!=y))
print("{0}=={1} la {2}".format(x,y,x==y))
print(f"{x}<{y} va {y}<{z} la {(x<y) and (y<z)}")
print((x<y) or (y>z))
print(not (x<y))
```
