Bài 13: Nhập vào số nguyên dương A, tìm n nhỏ nhất sao cho
1 + 1/2 + 1/3 + 1/4 + ... + 1/n > A 


```python
#bai 13
a = int(input())
sum = 0
i = 1
while sum<=a:
    sum = sum + 1/i
    i = i+1
    if sum >a:
        k = i
print(k)
```

Bài 14: Nhập vào một dãy số nguyên, dừng nhập khi người dùng nhập -1. Sau khi nhập
xong, in số lớn nhất, số nhỏ nhất trong những số vừa nhập. KHÔNG ĐƯỢC dùng mảng

```python
#bai14
n = float(input())
so_be_nhat = 100000000000000
while n != -1:
    if so_be_nhat>n:
        so_be_nhat=n
    else:
        so_be_nhat = so_be_nhat
    if n == -1:
        so_be_nhat = so_be_nhat
        break
    n = float(input())
print(so_be_nhat)
```

Bài 16: Nhập vào số nguyên dương n, đếm xem n có bao nhiêu chữ số chẵn, bao nhiêu
chữ số lẻ 

```python
#bai 16
n = int(input())
even = 0
odd = 0
while n>0:
    if n//10>0:
        if (n%10)%2==0:
            even = even +1
            n = n//10
        elif (n%10)%2!=0:
            odd = odd+1
            n = n//10
    if n//10==0:
                if (n%10)%2==0:
                    even = even +1
                    break
                elif (n%10)%2!=0:
                    odd = odd+1
                    break
print (odd)
print (even)
```

Bài 19: Dãy số fibonacci là dãy số được định nghĩa như sau: 1, 1, 2, 3, 5, 8, 13,... với số
kế tiếp sẽ bằng tổng hai số trước đó. Nhập vào số nguyên dương A, hãy tìm số trong
dãy số fibonacci lớn nhất nhưng không vượt quá A

```python
#bai 19
n = int(input())
sum1 = 0
sum2 = 1
while sum2+sum1<n:
    sum2 = sum1 + sum2
    if sum2>n:
        sum2 = sum2+sum1
        sum1 = sum2 -sum1
        sum = sum2 - sum1
        break
    sum1 = sum2 -sum1
print(sum)
```
