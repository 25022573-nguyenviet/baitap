```python
#W4A18
a,b = map(int,input().split())
for i in range(1,a+1):
    if a%i==0 and b%i==0:
        print(i,end=" ")
```

W4A25
Nhập vào một dãy số nguyên, ngưng nhập khi người dùng nhập -1.
Sau khi nhập xong, in số lớn nhất, số nhỏ nhất trong những số vừa nhập 


```python
#W4A25
n = int(input())
so_be_nhat = 99999999999999999999
so_lon_nhat = -1000000000000000000000
while n!=-1:
    if so_be_nhat>=n:
        so_be_nhat = n
    else:
        pass
    if so_lon_nhat<=n:
        so_lon_nhat = n
    else:
        pass
    n = int(input())
print (so_lon_nhat , so_be_nhat)
    
```

[Fibonaci1]
Dãy số fibonacci là dãy số được định nghĩa như sau: 1, 1, 2, 3, 5, 8, 13,... với số kế tiếp sẽ
bằng tổng hai số trước đó
Nhập vào số nguyên dương A, hãy tìm số trong dãy số fibonacci lớn nhất nhưng không vượt
quá A 


```python
#W4A26
n = int(input())
sum1 = 0
sum2 = 1
sum3 = sum1 + sum2
while sum3 <n:
    sum3 = sum1 + sum2
    sum1 = sum2
    sum2 = sum3
else:
    print(sum1)
```
