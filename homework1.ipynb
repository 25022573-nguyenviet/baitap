9. Chuyển đổi số thành chữ số: Nhập một số nguyên từ 0 đến 9, in ra tên chữ số
tương ứng (0 -> "Không", 1 -> "Một", ...) – Xem trước match …case 


```python
#bai 9
a = int(input())
match a:
    case 0:
        print ("khong")
    case 1:
        print ("mot")
    case 2:
        print ("hai")
    case 3:
        print ("ba")
    case 4:
        print ("bon")
    case 5:
        print ("nam")
    case 6:
        print ("sau")
    case 7:
        print ("bay")
    case 8:
        print ("tam")
    case 9:
        print ("chin")
    case _:
        print ("chi nhap tu 0-9")
```
